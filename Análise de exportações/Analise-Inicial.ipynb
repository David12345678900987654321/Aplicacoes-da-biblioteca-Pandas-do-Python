### Como foi a evolução das exportações para a frança ao longo dos anos?

Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
Exportacoes_df = Tabela_df[["Year","US$ FOB"]]
Exportacoes_df= Exportacoes_df.groupby("Year").sum(numeric_only=True)
display(Exportacoes_df)
```

### Quais os produtos mais exportados ao longo de todo o período?

Queremos saber os produtos com maior valor US$ de exportação

```python
Produtos_exportados_df = Tabela_df[["SH2 Description","US$ FOB"]]
Produtos_exportados_df = Produtos_exportados_df.groupby("SH2 Description").sum(numeric_only=True)
Produtos_exportados_df = Produtos_exportados_df.sort_values(by = "US$ FOB", ascending=False)
display(Produtos_exportados_df)
```

### Em 2020 qual cidade mais exportou para a França?
Em valores US$

```python
Exportacoes2020 = Tabela_df["Year"] == 2020
Exportacoes2020_df = Tabela_df[Exportacoes2020]
Exportacoes2020Filtro_df = Exportacoes2020_df[["City","US$ FOB"]]
Exportacoes2020Filtro_df = Exportacoes2020Filtro_df.groupby("City", as_index=False).sum(numeric_only=True)
Exportacoes2020Filtro_df = Exportacoes2020Filtro_df.sort_values(by="US$ FOB", ascending=False)
display(Exportacoes2020Filtro_df)
print(f"Cidade que mais Exportou para a França {Exportacoes2020Filtro_df.iloc[0]["City"]}")
```
